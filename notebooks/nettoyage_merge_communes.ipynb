# Nettoyage et merge des bases de données des communes

Merge les bases sur `codgeo` et `an`.
La base finale a une ligne par couple **(commune × année)**.

```python
import pandas as pd
import os
```

## 1. Configuration

```python
# Adapter ce chemin à votre environnement local
base_path = "/Users/margo/Documents/Margot/0.Polytechnique/Statistical-machine-learning/Projet-final/Data/"
obs_path = os.path.join(base_path, "Observatoire-territoires")
```

## 2. Chargement et nettoyage de chaque base

**Structure commune à tous les fichiers :**
- Lignes 0-3 : métadonnées ANCT (à ignorer)
- Ligne 4 : en-têtes réels (`codgeo`, `libgeo`, `an`, + variable)
- Lignes 5+ : données

Le nettoyage consiste à sauter les 4 premières lignes et à ne garder que les colonnes utiles.

```python
def load_obs_file(filepath):
    """
    Charge un fichier Excel de l'Observatoire des territoires.
    Ignore les 4 premières lignes de métadonnées.
    Retourne un DataFrame avec codgeo (str), libgeo, an, et la variable.
    """
    df = pd.read_excel(filepath, header=4)  # La ligne 4 (index) = vrais en-têtes
    df['codgeo'] = df['codgeo'].astype(str).str.zfill(5)  # Conserve le zéro initial (ex: 01001)
    return df
```

```python
# Chargement de toutes les bases
obs_data = {}
for file in sorted(os.listdir(obs_path)):
    if file.endswith(".xlsx"):
        name = file.replace(".xlsx", "")
        filepath = os.path.join(obs_path, file)
        obs_data[name] = load_obs_file(filepath)
        print(f"✓ {name} — {len(obs_data[name])} lignes, colonnes : {list(obs_data[name].columns)}")
```

## 3. Aperçu rapide de chaque base

```python
for name, df in obs_data.items():
    print(f"\n=== {name} ===")
    print(f"  Dimensions : {df.shape}")
    print(f"  Années disponibles : {sorted(df['an'].dropna().unique()) if 'an' in df.columns else 'N/A'}")
    display(df.head(3))
```

# 4. Séparation des bases à l'échelle communale et à l'échelle départementale
Il y a des bases de données qui ne sont pas à l'échelle de la commune mais du département. Une solution : attribuer la même valeur aux communes du même département. Une autre plus radicale, supprimer cette base de données. 

On pourra, après avoir pris une décision, supprimer le code chunk suivant (pas très utile)

```python
# Identifier les bases par niveau géographique
bases_commune = {}
bases_dept = {}

for name, df in obs_data.items():
    # Les codgeo départementaux ont 2 caractères (avant zfill(5) → '00001' pour dept '1')
    # On détecte en regardant la longueur des codgeo AVANT le zfill
    raw = pd.read_excel(os.path.join(obs_path, [f for f in os.listdir(obs_path) if name in f][0]), header=4)
    raw_codgeo = raw['codgeo'].astype(str).str.strip()
    max_len = raw_codgeo.dropna().apply(len).max()
    
    if max_len <= 3:  # Département ou région
        bases_dept[name] = df
        print(f"[DÉPARTEMENT] {name}")
    else:
        bases_commune[name] = df
        print(f"[COMMUNE]     {name}")
```

## 5. Séparation des bases sans colonne `an` (données transversales)
Même dilemme : est-ce qu'on rend constant à travers le temps ces variables ? Ou tout simplement on les supprime ?

A faire : regarder pour chaque base de données quelles sont les années présentes

```python
# Certaines bases n'ont pas de dimension temporelle (ex: FRR)
bases_avec_an = {}
bases_sans_an = {}

for name, df in bases_commune.items():
    if 'an' in df.columns and df['an'].notna().any():
        bases_avec_an[name] = df
    else:
        bases_sans_an[name] = df
        print(f"[SANS ANNÉE] {name} — sera jointe sur codgeo uniquement")

print(f"\nBases avec année : {len(bases_avec_an)}")
print(f"Bases sans année : {len(bases_sans_an)}")
```

## 6. Construction de la base finale (commune × année)

**Stratégie :** merge progressif sur `codgeo` + `an` avec `outer join` pour conserver toutes les communes et années présentes dans au moins une base.

```python
def get_variable_cols(df):
    """Retourne les colonnes qui ne sont pas codgeo, libgeo, an."""
    return [c for c in df.columns if c not in ['codgeo', 'libgeo', 'an']]

def load_obs_file(filepath):
    df = pd.read_excel(filepath, header=4)
    # Nettoyage du Code INSEE
    df['codgeo'] = df['codgeo'].astype(str).str.zfill(5)
    

    if 'an' in df.columns:
        # On supprime les lignes où l'année est vide avant conversion
        df = df.dropna(subset=['an']) 
        # On convertit en entier pour éviter les "2022.0"
        df['an'] = df['an'].astype(int) 
    
    return df

# Initialisation avec la première base ayant une colonne 'an'
base_names = list(bases_avec_an.keys())
merged = bases_avec_an[base_names[0]][['codgeo', 'libgeo', 'an'] + get_variable_cols(bases_avec_an[base_names[0]])].copy()
print(f"Base initiale : {base_names[0]} — {merged.shape}")

# Merge progressif
for name in base_names[1:]:
    df = bases_avec_an[name]
    var_cols = get_variable_cols(df)
    df_to_merge = df[['codgeo', 'an'] + var_cols].copy()
    
    merged = merged.merge(df_to_merge, on=['codgeo', 'an'], how='outer')
    
    # Récupérer libgeo si manquant
    if 'libgeo' not in merged.columns:
        merged = merged.merge(df[['codgeo', 'libgeo']].drop_duplicates(), on='codgeo', how='left')
    
    print(f"  + {name} → {merged.shape}")

print(f"\nBase mergée (avec années) : {merged.shape}")
```

```python
# Ajout des bases sans colonne 'an' (jointure sur codgeo uniquement)
for name, df in bases_sans_an.items():
    var_cols = get_variable_cols(df)
    df_to_merge = df[['codgeo'] + var_cols].drop_duplicates('codgeo')
    merged = merged.merge(df_to_merge, on='codgeo', how='left')
    print(f"  + {name} (sans an) → {merged.shape}")
```

```python
# Nettoyage final : réorganiser les colonnes
id_cols = ['codgeo', 'libgeo', 'an']
other_cols = [c for c in merged.columns if c not in id_cols]
merged = merged[id_cols + other_cols]

# Trier par commune puis année
merged = merged.sort_values(['codgeo', 'an']).reset_index(drop=True)

print(f"Base finale : {merged.shape[0]} lignes × {merged.shape[1]} colonnes")
merged.head(10)
```

## 7. Informations sur la base finale

```python
print("=== Résumé de la base finale ===")
print(f"Nombre de communes uniques : {merged['codgeo'].nunique()}")
print(f"Années couvertes : {sorted(merged['an'].dropna().unique())}")
print(f"\nVariables disponibles ({len(other_cols)}) :")
for col in other_cols:
    n_non_null = merged[col].notna().sum()
    pct = 100 * n_non_null / len(merged)
    print(f"  {col:<35} {n_non_null:>8} valeurs non-nulles ({pct:.1f}%)")
```

## 9. (Optionnel) Jointure avec la base département

La base `minimum_vieillesse` est au niveau département. On peut l'ajouter en faisant correspondre les 2 premiers caractères du `codgeo` commune.

```python
if bases_dept:
    merged['codgeo_dept'] = merged['codgeo'].str[:2]
    
    for name, df in bases_dept.items():
        var_cols = get_variable_cols(df)
        df_dept = df[['codgeo', 'an'] + var_cols].copy()
        df_dept = df_dept.rename(columns={'codgeo': 'codgeo_dept'})
        df_dept['codgeo_dept'] = df_dept['codgeo_dept'].astype(str).str.zfill(2)
        
        if 'an' in df_dept.columns:
            merged = merged.merge(df_dept, on=['codgeo_dept', 'an'], how='left')
        else:
            merged = merged.merge(df_dept.drop_duplicates('codgeo_dept'), on='codgeo_dept', how='left')
        
        print(f"  + {name} (département) → {merged.shape}")
    
    merged = merged.drop(columns=['codgeo_dept'])
    print(f"\nBase finale avec données département : {merged.shape}")
    display(merged.head())
```
